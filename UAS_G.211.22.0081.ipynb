```python
import matplotlib.pyplot as plt
import numpy as np

# Data
tahun = ['1981-1990', '1991-2000', '2001-2010']
bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Agu', 'Sep', 'Okt', 'Nov', 'Des']
curah_hujan = [
    [480, 310, 295, 275, 200, 160, 40, 80, 140, 220, 420, 400],  # Curah hujan tahun 1981-1990
    [460, 420, 420, 350, 200, 160, 60, 70, 70, 400, 500, 410],  # Curah hujan tahun 1991-2000
    [420, 400, 380, 390, 120, 100, 50, 10, 70, 330, 640, 420]   # Curah hujan tahun 2001-2010
]

# Plot bar untuk setiap tahun
fig, ax = plt.subplots()
width = 0.2
offset = np.arange(len(bulan))

for i, t in enumerate(tahun):
    ax.bar(offset + i * width, curah_hujan[i], width, label=str(t))


# Penamaan sumbu dan judul
ax.set_xlabel('Bulan')
ax.set_ylabel('Curah Hujan (mm)')
ax.set_title('Curah Hujan per Bulan dan Tahun')

# Tampilkan legenda
ax.legend()

# Atur label sumbu x
ax.set_xticks(offset + (len(tahun) - 1) * width / 2)
ax.set_xticklabels(bulan)

plt.ylim(0, 700)

# Tampilkan plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Mengatur lebar plot
plt.figure(figsize=(10, plt.rcParams["figure.figsize"][1]))

# Memberikan judul dan label sumbu
plt.title('Suhu Permukaan')
plt.xlabel('\nWaktu')
plt.ylabel('Celcius')

# Mengatur batas sumbu y
plt.ylim(20, 27)

# Data suhu permukaan
waktu = ['00:00 UTC', '03:00 UTC', '06:00 UTC', '09:00 UTC', '12:00 UTC\n24 Juli 2013', '15:00 UTC', '18:00 UTC', '21:00 UTC', '00.00 UTC\n25 Juli 2013']
suhu_asimilasi = [23.1, 25.4, 25.6, 24.8, 22.5, 22.4, 23.9, 22, 25.7]
suhu_observasi = [24.1, 25.1, 23.8, 23.9, 24, 24.2, 24, 22.9, 24]
suhu_non_asimilasi = [23.4, 25.5, 25.3, 25.4, 24.9, 25.1, 24.7, 24.2, 25.9]

# Membuat plot menggunakan garis
plt.plot(waktu, suhu_asimilasi, label='Asimilasi', marker='o', color='blue')
plt.plot(waktu, suhu_observasi, label='Observasi', marker='o', color='yellow')
plt.plot(waktu, suhu_non_asimilasi, label='Non-Asimilasi', marker='o', color='red')

# Menampilkan legend
plt.legend(bbox_to_anchor=(1.10, 0.5), loc='upper center')

# Menampilkan plot
plt.show()

```
